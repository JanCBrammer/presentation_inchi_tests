```python
from tucan.graph_utils import permute_molecule
from tucan.io import graph_from_file
import networkx as nx
from pathlib import Path
import matplotlib.pyplot as plt

def draw_molecule(m):
    highlight_colors = list(nx.get_node_attributes(m, "atomic_number").values())
    node_size = 1 / m.order() * 10000
    fig, ax = plt.subplots(figsize=(7, 7))
    nx.draw_kamada_kawai(
        m,
        node_color=highlight_colors,
        node_size=node_size,
        width=5,
        cmap="rainbow",
        alpha=1,
        with_labels=True,
        font_weight="heavy",
        font_color="w",
        font_size=40,
        ax=ax,
    )
```

```python
molecule = "water"
m = graph_from_file(Path(f"{molecule}.mol"))
draw_molecule(m)
plt.savefig("water.svg", transparent=True)
```

```python
molecule = "BF3"
m = graph_from_file(Path(f"{molecule}.mol"))
draw_molecule(m)
plt.savefig("BF3.svg", transparent=True)

m = permute_molecule(m, random_seed=.41)
draw_molecule(m)
plt.savefig("BF3_permuted_1.svg", transparent=True)

m = permute_molecule(m, random_seed=.42)
draw_molecule(m)
plt.savefig("BF3_permuted_2.svg", transparent=True)
```
